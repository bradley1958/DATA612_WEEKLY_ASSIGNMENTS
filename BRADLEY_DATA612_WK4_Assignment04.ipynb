When we concatenate DateFrames, we need to specify the axis (axis=0) tells pandas to stack the second DataFrame on the first one. Pandas will automatically detect whether the column names are the same and will stack the data accordingly. Using (axis=1) will stack the columns in the second DataFrame to the "right" of the first DataFrame. To stack the data vertically, we need to make sure we have the same columns and associated column format in both datasets. When we stack horizontally, we want to make sure what we are doing makes sense (that the data points/elements are related in some way ) this ensures value with respect to the merged dataset.

```
# Let's place the two DataFrames side by side
horizontal_stack = pd.concat([df, stocks], axis=1)
horizontal_stack.info()
horizontal_stack.head(3)
```

As you can see from the above syntax we have merged our two datasets (df & stocks) into a new DataFrame called (horizontal_stack). 

```
horizontal_stack.isnull()
```

From our syntax above you can see by the output the we have many missing values denoted by "True" in the column output values.
Next, we will count the missing values.


```
# Let's count the missing values in our new "horizontal_stack" DataFrame
horizontal_stack.isnull().sum()
```

So, we have many missing values associated with what had been the "stocks" dataset that is now part of our merged " horizontal_stack" new dataset.


Next, we need to handle all the NaN values in our newly combined "horizontal_stack" dataset. Do we do this by replacing the NaN value with "zero", or do we use the "mean" to fill in the gaps in the data values. 

```
# Let's see what happens if we replace NaN with "zero" using "fillna"
horizontal_stack = horizontal_stack.fillna(0)
```

```
horizontal_stack.isnull().sum()
```

So, as you can see from the above syntax all the NaN values in our mereged dataset have been filled with "zero". All counts have been changed to "zero".
